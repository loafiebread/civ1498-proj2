# Bike Share Trip Data

```python
# Create a list of file names for bike share trips data
trips_filenames2017 = [filename for filename in os.listdir() if 'bike_share_2017' in filename]
trips_filenames2018 = [filename for filename in os.listdir() if 'bike_share_2018' in filename]
trips_filenames2019 = [filename for filename in os.listdir() if 'bike_share_2019' in filename]
trips_filenames2020 = [filename for filename in os.listdir() if 'bike_share_2020' in filename]
trips_filenames2020.remove('bike_share_2020-11.csv')
trips_filenames2020.remove('bike_share_2020-12.csv')
trips_filenames2020pt2 = ['bike_share_2020-11.csv','bike_share_2020-12.csv']
trips_filenames2021 = [filename for filename in os.listdir() if 'bike_share_2021' in filename]

trips_data_2017 =pd.concat([pd.read_csv(fname) for fname in trips_filenames2017],axis=0,ignore_index= True)
trips_data_2018 =pd.concat([pd.read_csv(fname) for fname in trips_filenames2018],axis=0,ignore_index= True)
trips_data2019 =pd.concat([pd.read_csv(fname) for fname in trips_filenames2019],axis=0,ignore_index= True)
trips_data2021 =pd.concat([pd.read_csv(fname) for fname in trips_filenames2021],axis=0,ignore_index= True)
trips_data2020pt1 =pd.concat([pd.read_csv(fname) for fname in trips_filenames2020],axis=0,ignore_index= True)
trips_data2020pt2 =pd.concat([pd.read_csv(fname) for fname in trips_filenames2020pt2],axis=0,ignore_index= True)

trips_data_2017.columns = [' '.join(col.split()) for col in trips_data_2017.columns]
trips_data_2018.columns = [' '.join(col.split()) for col in trips_data_2018.columns]
trips_data2019.columns = [' '.join(col.split()) for col in trips_data2019.columns]
trips_data2020pt1.columns = [' '.join(col.split()) for col in trips_data2020pt1.columns]
trips_data2020pt2.columns = [' '.join(col.split()) for col in trips_data2020pt2.columns]
trips_data2021.columns = [' '.join(col.split()) for col in trips_data2021.columns]
```

```python
trips_data_2017['trip_start_time'] = pd.DatetimeIndex(pd.to_datetime(trips_data_2017['trip_start_time'],format="%Y-%m-%d %H:%M (%Z)"))
trips_data_2017['trip_stop_time'] = pd.DatetimeIndex(pd.to_datetime(trips_data_2017['trip_stop_time'],format="%Y-%m-%d %H:%M (%Z)"))
trips_data_2018['trip_start_time'] = pd.DatetimeIndex(pd.to_datetime(trips_data_2018['trip_start_time'],format="%Y-%m-%d %H:%M (%Z)"))
trips_data_2018['trip_stop_time'] = pd.DatetimeIndex(pd.to_datetime(trips_data_2018['trip_stop_time'],format="%Y-%m-%d %H:%M (%Z)"))
trips_data_2018['trip_start_time'] = trips_data_2018['trip_start_time'].dt.tz_convert('EST')
trips_data_2018['trip_stop_time'] = trips_data_2018['trip_stop_time'].dt.tz_convert('EST')
trips_data_2017['trip_start_time'] = trips_data_2017['trip_start_time'].dt.tz_convert('EST')
trips_data_2017['trip_stop_time'] = trips_data_2017['trip_stop_time'].dt.tz_convert('EST')
```

```python
trips_data2019['Start Time'] = pd.DatetimeIndex(pd.to_datetime(trips_data2019['Start Time'],format="%d/%m/%Y %H:%M:%S (%Z)"))
trips_data2019['End Time'] = pd.DatetimeIndex(pd.to_datetime(trips_data2019['End Time'],format="%d/%m/%Y %H:%M:%S (%Z)"))
trips_data2020pt1['Start Time'] = pd.DatetimeIndex(pd.to_datetime(trips_data2020pt1['Start Time'],format="%d/%m/%Y %H:%M:%S (%Z)"))
trips_data2020pt1['End Time'] = pd.DatetimeIndex(pd.to_datetime(trips_data2020pt1['End Time'],format="%d/%m/%Y %H:%M:%S (%Z)"))
trips_data2020pt2['Start Time'] = pd.DatetimeIndex(pd.to_datetime(trips_data2020pt2['Start Time'],format="%d/%m/%Y %H:%M:%S (%Z)"))
trips_data2020pt2['End Time'] = pd.DatetimeIndex(pd.to_datetime(trips_data2020pt2['End Time'],format="%d/%m/%Y %H:%M:%S (%Z)"))
trips_data2021['Start Time'] = pd.DatetimeIndex(pd.to_datetime(trips_data2021['Start Time'],format="%d/%m/%Y %H:%M:%S (%Z)"))
trips_data2021['End Time'] = pd.DatetimeIndex(pd.to_datetime(trips_data2021['End Time'],format="%d/%m/%Y %H:%M:%S (%Z)"))
```

```python
trips_data_2017 = trips_data_2017.rename(columns={'trip_id': 'Trip Id',
                                                'trip_start_time': 'Start Time',
                                                'trip_stop_time': 'End Time', 
                                                'trip_duration_seconds': 'Trip Duration',
                                                'from_station_id': 'Start Station Id', 
                                                'from_station_name': 'Start Station Name', 
                                                'to_station_id': 'End Station Id', 
                                                'to_station_name': 'End Station Name', 
                                                'user_type': 'User Type'})

trips_data_2018 = trips_data_2018.rename(columns={'trip_id': 'Trip Id',
                                                'trip_start_time': 'Start Time',
                                                'trip_stop_time': 'End Time', 
                                                'trip_duration_seconds': 'Trip Duration',
                                                'from_station_id': 'Start Station Id', 
                                                'from_station_name': 'Start Station Name', 
                                                'to_station_id': 'End Station Id', 
                                                'to_station_name': 'End Station Name', 
                                                'user_type': 'User Type'})
```

```python
df = pd.concat([trips_data_2017,trips_data_2018, trips_data2019,trips_data2020pt1,trips_data2020pt2,trips_data2021])
df = df.drop(columns=['ï»¿Trip Id','Subscription Id','Bike Id'])
```

```python
df=df.replace('Member','Annual Member')
df=df.replace('Casual','Casual Member')
```

```python
from scipy import stats
df.dropna()
df = df[(np.abs(stats.zscore(stats.zscore(df['Trip Duration']))) < 3)]
```

```python
start_stations = pd.read_csv('bikeshare_stations.csv')
start_stations = start_stations.rename(columns={"name": "Start Station Name", "lat": "Start_Lat", "lon": "Start_Lon"})
end_stations = pd.read_csv('bikeshare_stations.csv')
end_stations = end_stations.rename(columns={"name": "End Station Name", "lat": "End_Lat", "lon": "End_Lon"})
```

```python
df = pd.merge(df, start_stations, on='Start Station Name')
df = pd.merge(df, end_stations, on='End Station Name')
df.head()
```

```python
df = df.drop(columns=['station_id_x', 'station_id_y'])
df.info()
```

```python
def calc_distance(lat_from, lon_from, lat_to, lon_to):
    coords_1 = (lat_from, lon_from)
    coords_2 = (lat_to, lon_to)
    distance = geopy.distance.distance(coords_1, coords_2).m
    
    return distance
```

```python
df['Distance'] = df.apply(lambda x: calc_distance(x['Start_Lat'],x['Start_Lon'],x['End_Lat'],x['End_Lon']),axis=1)
```

```python
df.to_csv('trips_with_coordinates_distance.csv', index=False)
```
